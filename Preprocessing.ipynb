# Handling Missing values

```python
def missing_values(df):
    dic = {'ColumnName':[], 'MissingValues':[], 'Percentage':[]}
    for col in df.columns:
        dic['ColumnName'].append(col)
        null = df[col].isna().sum()
        dic['MissingValues'].append(null)
        dic['Percentage'].append((null/df.shape[0])*100)
    return pd.DataFrame(dic).sort_values('Percentage', ascending= False)

```

```python
def fill_city(city, incident):
    if type(city)== float:
        return incident
    else:
        return city
insurance_df['CITY']=insurance_df.apply(lambda row: fill_city(row['CITY'], row['INCIDENT_CITY']), axis = 1)
```

```python
def fill_incident_city(city, incident):
    if type(incident)== float:
        return city
    else:
        return incident
insurance_df['INCIDENT_CITY']=insurance_df.apply(lambda row: fill_incident_city(row['CITY'], row['INCIDENT_CITY']), axis = 1)
```

```python
vendor_ratio = insurance_df['VENDOR_ID'].value_counts(normalize=True)

def fill_missing_vendor_ids(df, column_name, ratio):
    # Get the indices where the column has NaN values
    missing_indices = df[column_name].isna()
    
    # Use np.random.choice to randomly select a value based on the ratio
    df.loc[missing_indices, column_name] = np.random.choice(
        ratio.index, size=missing_indices.sum(), p=ratio.values
    )

# Apply the function to fill missing values in 'VENDOR_ID' column
fill_missing_vendor_ids(insurance_df, 'VENDOR_ID', vendor_ratio)
```

```python
def fill_authority(insurance_type, authority):
    if type(authority)==float:
        if insurance_type in ['Life', 'Health', 'Travel']:
            return 'Ambulance'

        elif insurance_type in ['Motor', 'Property']:
            return 'Police'
        else:
            return 'Other'
    else:
        return authority
    

insurance_df.AUTHORITY_CONTACTED  =insurance_df.apply(lambda row: fill_authority(row['INSURANCE_TYPE'], row['AUTHORITY_CONTACTED']), axis = 1)
```
