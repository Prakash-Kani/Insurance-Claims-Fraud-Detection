```python
plt.figure(figsize= (8,8))
plt.pie(x = df['RISK_SEGMENTATION'].value_counts(), labels=['LOW', 'MEDIUM', 'HIGH'], colors=['g', 'b', 'r'], explode = [0, 0.1,0.2 ], shadow = True, autopct='%1.1f%%')
plt.title('Distribution of Risk Segmentations')
plt.show()
```

Certainly! Here are some exploratory data analysis (EDA) insights based on the provided pie chart depicting the distribution of risk segmentations: ### EDA Insights: 1. **Overall Distribution**: - The pie chart represents the distribution of risk segmentations into three categories: LOW, MEDIUM, and HIGH. - The percentages for each category are as follows: - LOW: 44.0% - MEDIUM: 41.5% - HIGH: 14.6% 2. **Dominant Risk Segment**: - The LOW risk segment is the largest category, comprising 44.0% of the total distribution. This indicates that nearly half of the data points fall into the LOW risk category. 3. **Medium Risk Segment**: - The MEDIUM risk segment is the second largest category, making up 41.5% of the total distribution. This is very close to the LOW risk segment, suggesting a significant portion of the data points are also in the MEDIUM risk category. 4. **High Risk Segment**: - The HIGH risk segment is the smallest category, accounting for only 14.6% of the total distribution. This indicates that a relatively small portion of the data points are considered to be in the HIGH risk category. 5. **Comparative Analysis**: - The LOW and MEDIUM risk segments together account for 85.5% of the total distribution, which is a substantial majority. This suggests that the majority of the data points are not in the HIGH risk category. - The HIGH risk segment is significantly smaller compared to the other two segments, indicating that high-risk cases are less common in the dataset. 6. **Implications**: - The distribution suggests that the majority of the subjects or items being analyzed are at a lower or medium risk, which could be a positive indicator depending on the context (e.g., health, finance, etc.). - The relatively small percentage of high-risk cases might warrant focused attention to mitigate potential adverse outcomes. 7. **Visual Representation**: - The use of distinct colors (green for LOW, blue for MEDIUM, and red for HIGH) helps in quickly identifying and differentiating between the risk segments. - The 3D effect and the slight separation of the segments enhance the visual appeal and clarity of the chart. ### Conclusion: The pie chart provides a clear and concise visual representation of the distribution of risk segmentations. The insights derived from this chart can help in understanding the overall risk profile and in making informed decisions based on the distribution of risk categories.

```python
df1 = df.copy()
```

```python
# Define risk segmentation based on CLAIM_AMOUNT thresholds
bins = [0, 5000, 20000, df1['CLAIM_AMOUNT'].max()]
labels = ['Low Risk', 'Medium Risk', 'High Risk']
df1['RISK_SEGMENT'] = pd.cut(df1['CLAIM_AMOUNT'], bins=bins, labels=labels)

# Plot the distribution of Risk Segmentation
plt.figure(figsize=(10, 6))
sns.countplot(x='RISK_SEGMENT', data=df1, palette='Set2')
plt.title('Distribution of Risk Segmentation')
plt.xlabel('Risk Segment')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot to compare PREMIUM_AMOUNT across risk segments
plt.figure(figsize=(10, 6))
sns.boxplot(x='RISK_SEGMENTATION', y='PREMIUM_AMOUNT', data=df, palette='Set3')
plt.title('Premium Amount by Risk Segment')
plt.xlabel('Risk Segment')
plt.ylabel('Premium Amount')
plt.show()
```

```python

# Group by CITY and get normalized value counts for RISK_SEGMENTATION
normalized_counts = df.groupby('CITY')['RISK_SEGMENTATION'].value_counts(normalize=True).unstack()

# Get the top 10 cities based on the total normalized counts
top_cities = normalized_counts.sum(axis=1).nlargest(10).index
top_normalized_counts = normalized_counts.loc[top_cities]

# Plotting using seaborn
plt.figure(figsize=(12, 6))
top_normalized_counts.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('City')
plt.ylabel('Normalized Risk Segmentation Count')
plt.title('Top 10 Cities by Risk Segmentation (Normalized)')

# Display the plot
plt.xticks(rotation=45)
plt.legend(title='Risk Segmentation')
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```
